```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Lê o arquivo CSV
df = pd.read_csv(
    "https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/alunos_cursos.csv",
    sep=";"
)

# Filtros de cursos
var_curso = "Excel"
var_curso2 = "Python"

# Filtra somente os cursos desejados
df_filtrado = df[(df["Curso"] == var_curso) | (df["Curso"] == var_curso2)]

# Gráfico
sns.lineplot(x="Aluno", y="Nota", data=df_filtrado, marker="o", hue="Curso")
plt.title(f"Média de notas por curso - {var_curso} e {var_curso2}")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
